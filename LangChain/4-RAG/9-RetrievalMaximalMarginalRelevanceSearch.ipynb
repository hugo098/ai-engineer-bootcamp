```python
import os
from dotenv import load_dotenv

load_dotenv()

api_key = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai.embeddings import OpenAIEmbeddings
from langchain_core.documents import Document
from langchain_chroma import Chroma
```

```python
embeddings = OpenAIEmbeddings(openai_api_key=api_key, model="text-embedding-ada-002")
```

```python
vectorstore_from_directory = Chroma(
    persist_directory="./chroma_data_science_course",
    embedding_function=embeddings,
    collection_name="data_science_course",
)
```

```python
question = "What software do data scientists use?"
```

```python
retrieved_docs = vectorstore_from_directory.max_marginal_relevance_search(
    query=question,
    k=3,
    lambda_mult=0.7,
    filter={
        "Lecture Title": {
            "$eq": "Programming Languages & Software Employed in Data Science - All the Tools You Need"
        }
    },
)
```

```python
for i in retrieved_docs:
    print(i.page_content)
    print(i.metadata)
    print("-----")
```
