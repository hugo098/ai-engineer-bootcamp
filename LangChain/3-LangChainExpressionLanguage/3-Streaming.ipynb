### Streaming


```python
import os
from dotenv import load_dotenv

load_dotenv()
api_key = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
```

```python
chat_template = ChatPromptTemplate.from_messages(
    [
        (
            "human",
            "I have recently adopted a {pet} which is a {breed}. Coul you suggest several training tips?",
        )
    ]
)
```

```python
chat = ChatOpenAI(
    api_key=api_key,
    model_name="gpt-4.1-mini",
    temperature=0,
    max_tokens=100,
    seed=42,
)
```

```python
chain = chat_template | chat
```

```python
chain.invoke({"pet": "dog", "breed": "Golden Retriever"})
```

```python
response = chain.stream({"pet": "cat", "breed": "Siamese"})
```

```python
next(response)
```

```python
for i in response:
    print(i.content, end="")
```
