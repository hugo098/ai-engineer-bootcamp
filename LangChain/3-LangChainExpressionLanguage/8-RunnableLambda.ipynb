### RunnableLambda


```python
from langchain_core.runnables import RunnableLambda
```

```python
find_sum = lambda numbers: sum(numbers)
```

```python
find_sum([1, 2, 3, 4, 5])
```

```python
find_square = lambda n: n**2
```

```python
find_square(8)
```

```python
runnable_sum = RunnableLambda(lambda numbers: sum(numbers))
```

```python
runnable_sum.invoke([1, 2, 3, 4, 5])
```

```python
runnable_square = RunnableLambda(lambda n: n**2)
```

```python
runnable_square.invoke(8)
```

```python
chain = runnable_sum | runnable_square
```

```python
chain.invoke([1, 2, 3, 4, 5])
```

```python
chain.get_graph().print_ascii()
```
