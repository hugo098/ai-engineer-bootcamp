### The @chain Decorator


```python
from langchain_core.runnables import RunnableLambda
from langchain_core.runnables import chain
```

```python
def find_sum(x):
    return sum(x)


def find_square(x):
    return x**2
```

```python
chain1 = RunnableLambda(find_sum) | RunnableLambda(find_square)
```

```python
chain1.invoke([1, 2, 3, 4, 5])
```

```python
@chain
def runnable_sum(x):
    return sum(x)


@chain
def runnable_square(x):
    return x**2
```

```python
chain2 = runnable_sum | runnable_square
```

```python
chain2.invoke([1, 2, 3, 4, 5])
```
