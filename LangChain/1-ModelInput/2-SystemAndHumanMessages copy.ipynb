```python
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
api_key = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.messages import HumanMessage, SystemMessage
```

```python
chat = ChatOpenAI(
    api_key=api_key,
    model_name="gpt-4.1-mini",
    temperature=0,
    max_tokens=100,
    seed=42,
)
```

```python
message_s = SystemMessage(content="You are a sarcastic assistant that tells jokes.")
message_h = HumanMessage(content="Tell me what is programming.")
```

```python
response = chat.invoke([message_s, message_h])
```

```python
print(response.content)
```
