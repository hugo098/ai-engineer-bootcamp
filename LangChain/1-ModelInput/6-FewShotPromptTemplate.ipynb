```python
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
api_key = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.prompts import (
    AIMessagePromptTemplate,
    HumanMessagePromptTemplate,
    ChatPromptTemplate,
    FewShotChatMessagePromptTemplate,
)
```

```python
chat = ChatOpenAI(
    api_key=api_key,
    model_name="gpt-4.1-mini",
    temperature=0,
    max_tokens=100,
    seed=42,
)
```

```python
TEMPLATE_H = """I've recently adopted a {pet}.
 Can you suggest some {pet} names?"""

TEMPLATE_AI = """{response}"""

message_template_h = HumanMessagePromptTemplate.from_template(TEMPLATE_H)
message_template_ai = AIMessagePromptTemplate.from_template(TEMPLATE_AI)
```

```python
example_template = ChatPromptTemplate.from_messages(
    [message_template_h, message_template_ai]
)
```

```python
examples = [
    {"pet": "dog", "response": "Here are some dog names: Max, Bella, Charlie."},
    {"pet": "cat", "response": "Here are some cat names: Luna, Oliver, Simba."},
]
```

```python
few_shot_prompt = FewShotChatMessagePromptTemplate(
    examples=examples,
    example_prompt=example_template,
)
```

```python
chat_template = ChatPromptTemplate.from_messages([few_shot_prompt, message_template_h])
```

```python
chat_value = chat_template.invoke({"pet": "rabbit"})
```

```python
chat_value
```

```python
for i in chat_value.messages:
    print(f"{i.type}: {i.content}")
```

```python
response = chat.invoke(chat_value)
```
