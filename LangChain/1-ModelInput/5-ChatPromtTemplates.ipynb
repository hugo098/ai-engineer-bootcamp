```python
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
api_key = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.prompts.chat import (
    SystemMessagePromptTemplate,
    HumanMessagePromptTemplate,
    ChatPromptTemplate,
)
```

```python
chat = ChatOpenAI(
    api_key=api_key,
    model_name="gpt-4.1-mini",
    temperature=0,
    max_tokens=100,
    seed=42,
)
```

```python
TEMPLATE_S = "{description}"
TEMPLATE_H = "I've recently adopted a {pet}. Could you suggest some {pet} names?"
```

```python
message_template_system = SystemMessagePromptTemplate.from_template(TEMPLATE_S)
message_template_human = HumanMessagePromptTemplate.from_template(TEMPLATE_H)
```

```python
message_template_system
```

```python
chat_template = ChatPromptTemplate.from_messages(
    [message_template_system, message_template_human]
)
```

```python
chat_template
```

```python
chat_value = chat_template.invoke(
    {
        "description": "You are a helpful assistant that helps people find pet names.",
        "pet": "dog",
    }
)
```

```python
chat_value
```

```python
response = chat.invoke(chat_value)
```
