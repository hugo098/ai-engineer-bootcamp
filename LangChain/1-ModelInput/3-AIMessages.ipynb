```python
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
api_key = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.messages import HumanMessage, AIMessage
```

```python
chat = ChatOpenAI(
    api_key=api_key,
    model_name="gpt-4.1-mini",
    temperature=0,
    max_tokens=100,
    seed=42,
)
```

```python
message_h = HumanMessage(content="Tell me what is programming.")
message_ai = AIMessage(
    content="Oh, you know, just the ancient art of talking to rocks we tricked into thinking by feeding them lightning. Super simple stuff."
)
message_h2 = HumanMessage(content="What is radiation?")
message_ai2 = AIMessage(
    content="Radiation is like the universe's way of sending out invisible postcardsâ€”whether it's from the sun, radioactive materials, or even your microwave oven. Just don't try to read them without protection!"
)
message_h3 = HumanMessage(content="what is life?")
message_ai3 = AIMessage(
    content="Oh, just a brief, cosmic accident where carbon-based organisms frantically search for meaning while paying taxes and avoiding death. You know, the usual."
)
message_h4 = HumanMessage(content="What is hidrogen?")
```

```python
response = chat.invoke(
    [
        message_h,
        message_ai,
        message_h2,
        message_ai2,
        message_h3,
        message_ai3,
        message_h4,
    ]
)
```
