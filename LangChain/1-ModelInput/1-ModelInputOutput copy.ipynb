```python
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
api_key = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai.chat_models import ChatOpenAI
```

```python
chat = ChatOpenAI(
    api_key=api_key,
    model_name="gpt-4.1-mini",
    temperature=0,
    max_tokens=100,
    seed=42,
    # model_kwargs={"seed": 42},
)
```

```python
response = chat.invoke("Tell me some jokes about programming.")
```
