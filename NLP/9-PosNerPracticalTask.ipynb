```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import spacy
import re
import pandas as pd
import matplotlib.pyplot as plt
```

## Load data


```python
bbc_data = pd.read_csv("bbc_news.csv")
```

```python
bbc_data.head()
```

```python
bbc_data.info()
```

```python
titles = pd.DataFrame(bbc_data["title"])
titles.head()
```

## Clean data


```python
# lowercase
titles["lowercase"] = titles["title"].str.lower()
titles.head()
```

```python
# stopwords removal
en_stopwords = set(stopwords.words("english"))
titles["no_stopwords"] = titles["lowercase"].apply(
    lambda x: " ".join([word for word in x.split() if word not in en_stopwords])
)
```

```python
# punctuation removal
titles["no_stopwords_nopunct"] = titles["no_stopwords"].apply(
    lambda x: re.sub(r"[^\w\s]", "", x)
)
```

```python
# tokenize
titles["tokens_raw"] = titles["title"].apply(word_tokenize)
titles["tokens_clean"] = titles["no_stopwords_nopunct"].apply(word_tokenize)
```

```python
titles.head()
```

```python
# lemmatization
lemmatizer = WordNetLemmatizer()
titles["tokens_clean_lemmatized"] = titles["tokens_clean"].apply(
    lambda tokens: [lemmatizer.lemmatize(token) for token in tokens]
)
titles.head()
```

```python
tokens_raw_list = sum(titles["tokens_raw"].tolist(), [])
tokens_clean_list = sum(titles["tokens_clean_lemmatized"].tolist(), [])
```

```python
tokens_raw_list
```

```python
tokens_clean_list
```

## POS Tagging


```python
nlp = spacy.load("en_core_web_sm")
```

```python
spacy_doc = nlp(" ".join(tokens_raw_list))
```

```python
pos_df = pd.DataFrame(
    [(token.text, token.pos_) for token in spacy_doc], columns=["token", "pos_tag"]
)
```

```python
pos_df.head()
```

```python
pos_df_counts = (
    pos_df.groupby(["token", "pos_tag"])
    .size()
    .reset_index(name="counts")
    .sort_values(by="counts", ascending=False)
)
pos_df_counts.head(10)
```

```python
nouns = pos_df_counts[pos_df_counts.pos_tag == "NOUN"].head(10)
nouns
```

```python
verbs = pos_df_counts[pos_df_counts.pos_tag == "VERB"].head(10)
verbs
```

```python
adjectives = pos_df_counts[pos_df_counts.pos_tag == "ADJ"].head(10)
adjectives
```
