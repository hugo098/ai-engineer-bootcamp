## Lowercasing


```python
sentence = "Test Hello World"
```

```python
lower = sentence.lower()
```

```python
print(lower)
```

```python
sentence_list = ["Hello WORLD", "SumMER", "WiNtEr"]
```

```python
lower_sentence_list = [sentence.lower() for sentence in sentence_list]
```

```python
print(lower_sentence_list)
```
